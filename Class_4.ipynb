```python
class Node:
    """
       This class has two purposes:
           A) Store some data.
           B) Point to the next node in a sequence of nodes.
    """
    
    def __init__(self, data):
        self.data = data
        self.next = None
        
    def __repr__(self):
        return "< Node %r >" % self.data
```

```python
class LinkedList:
    """
        The LinkedList class maintains the reference
            to the HEAD or BEGINNING of our list.
    """
    
    def __init__(self):
        self.head = None               # Note that when a new LL is intitalized,
                                       # self.head is None
         
    def append(self, data):
        if not self.head:              # If self.head is None (meaning we have a new list)
            self.head = Node(data)     # ...populate it with a Node (that contains data)
            print("Added (HEAD): %s" % self.head)
        else:
            tail = self.find_tail()    # First: find tail (remember: tail is a node that points to None)
            tail.next = Node(data)     # Set tail.next (attribute) to point to a new Node with data.
            print("Added (BODY): %s" % tail.next)
            
    def preppend(self, data):
        new_node = Node(data)
        new_node.next = self.head
        self.head = new_node
            
    def insert(self, position, data):
        """Insert a Node with 'data' AFTER a Node at position."""
        node_at_index = self.get_node_at_index(position)
        current_next_node = node_at_index.next
        new_node = Node(data)
        node_at_index.next = new_node
        new_node.next = current_next_node
            
    def get_node_at_index(self, target):
        current = self.head
        counter = 0
        while current:
            if counter == target:
                return current
            current = current.next
            counter += 1
        raise IndexError("Index out of bounds.")
        
    def index(self, value):
        current = self.head
        counter = 0
        while current:
            if current.data == value:
                return counter
            current = current.next
            counter += 1
        raise ValueError("%s is not in the list.")
        
    def remove_node(self, value):
        if self.head.data == value:
            self.head = self.head.next
        else:
            current = self.head
            previous = None
            counter = 0
            while current:
                previous = current
                current = current.next
                if current.data == value:
                    previous.next = current.next
                    break
        raise ValueError("%s is not in the list.")
            
    def find_tail(self):
        current = self.head            # Create a temporary variable with the reference to self.head
        while current.next:            # As long as current.next is NOT None.
            current = current.next     # Set current to point to current.next
        print("Found tail: %s" % current)
        return current                 # Finally, return current (which at this point has .next set to None)
    
    def search(self, target):
        current = self.head            # Step 1: Create a temp variable that points to HEAD.
        while current:                 # Step 2: Check if we've reached the tail (points to None)
            if current.data == target:
                return True
            print("Saw:  %s" % current)
            current = current.next     # Step 3: Continue looping by switching out "current" to ""
        return False
    
    def __str__(self):
        current = self.head
        temp_list = []
        while current:
            temp_list.append(str(current.data))
            current = current.next
        return "[%s]" % ", ".join(temp_list)
```

```python
my_ll = LinkedList()
for i in range(5):
    my_ll.append(i)
    
print(my_ll)
print("Is 4 in the linked list -> %s" % my_ll.search(4))
print("Is 5 in the linked list -> %s" % my_ll.search(5))
```

```python
class DNode:
    def __init__(self, data):
        self.data = data
        self.prev = None
        self.next = None
```

```python
class DLList:
    def __init__(self):
        self.head = None
    
    def insert(self, position, data):
        """Insert BEFORE a given position."""
        if not self.head:
            raise IndexError("Cannot find position in empty list.")
        if position == 0:
            new_node = DNode(data)
            new_node.next = self.head
            self.head = new_node
            new_node.next.prev = self.head
        counter = 1
        current = self.head.next
        while current:
            if counter == position:
                previous_node = current.prev
                new_node = DNode(data)
                previous_node.next = new_node
                new_node.prev = previous_node
                new_node.next = current
                current.prev = new_node
        raise IndexError("Index out of bounds.")
        
```

```python
dlist = DLList()
dlist.head = DNode("A")
dlist.insert(0, "B")
dlist.insert(0, "C")

current = dlist.head
while current:
    print(current.data)
    current = current.next
```
